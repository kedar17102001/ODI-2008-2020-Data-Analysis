```python
unique_cols =  [x for x in df1.columns if df1[x].nunique()==1] 
print(unique_cols)
df1.drop(unique_cols,axis=1,inplace=True)
df1.columns
```

```python
df1[['Player', 'Region']] = df1['Player'].str.split('(', n=1, expand=True)
```

```python
df1[['Start Year','Last Year']]=df1['Span'].str.split('-',n=1, expand=True)
```

```python
df1.head(2)
```

```python
df1['Region'] = df1['Region'].map(lambda x: x.rstrip(')'))
```

```python
df1[['Region1', 'Region']] = df1['Region'].str.split('/', n=1, expand=True)
df1[['Region2', 'Region']] = df1['Region'].str.split('/', n=1, expand=True)
```

```python
def new_co(df1):
    if df1['Region1'] is not None:
        if df1['Region1'].isupper():
            if (df1.Region1=='ICC'):
                return df1['Region2']
            else:
                return df1['Region1']
                
        elif df1['Region'] is not None:
            if df1['Region'].isupper():
                if df1['Region'] is not "ICC":
                    return df1['Region']
            else:
                return df1['Region2']
        else:
                return df1['Region2']
    else:
        return "NA"
```

```python
df1['Final Region'] = df1.apply(new_co, axis=1)
df1
```
